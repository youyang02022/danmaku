```python
from sklearn.datasets import load_breast_cancer
 #jeu de données dans scikitlearn sont généralement sous la forme d'objts Bunch
import numpy as np
cancer  = load_breast_cancer()
#print('cancer.keys():{}'.format(cancer.keys()))
#print('target：{}'.format (cancer.target))
#('target_names：{}'.format (cancer.target_names))
print((cancer.data))  #When the PLUS symbol is used it means the decimal point has been moved to the RIGHT

print('shape of cancer data:{}'.format(cancer.data.shape))
print('feature names:\n{}'.format(cancer.feature_names))
print('description:\n{}'.format(cancer.DESCR))
print('sample counts per class:\n{}'.format({n:v for n,v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test = train_test_split(
    cancer.data,cancer.target,stratify=cancer.target,random_state = 66)
clf = KNeighborsClassifier()
#print('Training data size:{}'.format(len(X_train)))
print('Test data size:{}'.format(len(X_test)))
clf.fit(X_train,y_train)
print('Test set predictions :{}'.format(clf.predict(X_test)))
print('Test labels:{}'.format(y_test) )
print('Test set accuracy:{:.2f}'.format(clf.score(X_test,y_test)))

training_accuracy  = []
test_accuracy = []
neighbors_settings = range(1,20)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))

    
    
```
